# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Import picture

```python
img = cv2.imread("audi.jpg")
plt.imshow(img)
```

# Fix Image

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Classifier

```python
plate_classifier = cv2.CascadeClassifier("haar-cascade-files-master/haarcascade_russian_plate_number.xml")
```

# MAGIC

```python
def detect_plate(img):
    plate_img = img.copy()
    
    plate_rects = plate_classifier.detectMultiScale(plate_img, 1.1, 1)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate_img,
                     (x,y),
                     (x+w,y+h),
                     (255,255,255),
                     5)
    return plate_img
```

# Result

```python
result = detect_plate(fix_img)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
plt.imshow(result)
```
